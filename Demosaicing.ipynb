```python
#import the libararies that I need

import os

import colour

from colour_demosaicing import (
    EXAMPLES_RESOURCES_DIRECTORY,
    demosaicing_CFA_Bayer_bilinear,
    demosaicing_CFA_Bayer_Malvar2004,
    demosaicing_CFA_Bayer_Menon2007,
    mosaicing_CFA_Bayer)

colour.plotting.colour_style()

colour.utilities.describe_environment();
```

```python
# Read the original photo
LIGHTHOUSE_IMAGE = colour.io.read_image(os.path.join('photo', 'correct.jpg'))

# plotting the original photo
colour.plotting.plot_image(
    colour.cctf_encoding(LIGHTHOUSE_IMAGE),
    text_kwargs={'text': 'Lighthouse - R914108 - Kodak'});
```

```python
# mosaicing_CFA_Bayer is a function to return the Bayer CFA mosaic for a given RGB colourspace array or the original photo.
# A Bayer filter mosaic is a color filter array (CFA) for arranging RGB color filters on a square grid of photosensors.
CFA = mosaicing_CFA_Bayer(LIGHTHOUSE_IMAGE)

#plotting the photo after mosaicing
colour.plotting.plot_image(
    colour.cctf_encoding(CFA),
    text_kwargs={'text': 'Lighthouse - CFA - RGGB'})
```

```python
# Demosaicing is a digital image process used to reconstruct a full color image from the incomplete color samples output from an image sensor overlaid with a color filter array (CFA). 
# demosaicing_CFA_Bayer_Menon2007 is a function to return the demosaiced RGB colourspace array from given Bayer CFA using Menon (2007) demosaicing algorithm.

#plotting the photo after demosaicing
colour.plotting.plot_image(
    colour.cctf_encoding(demosaicing_CFA_Bayer_Menon2007(CFA)), 
    text_kwargs={'text': 'Lighthouse - Demosaicing - Menon (2007)'});
```
